```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np 
x = np.arange(0,100)
y = x*2
z= x**2
```

```python
fig = plt .figure()
ax =fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x_Axis')
ax.set_ylabel('y_Axis')
ax.set_title('Title')
```

```python
fig = plt .figure()
ax1 =fig.add_axes([0,0,1,1])
ax2 =fig.add_axes([0.2,0.5,0.2 ,0.2])
```

```python
fig = plt .figure()
ax1 =fig.add_axes([0,0,1,1])
ax1.plot(x,y ,color ='red')
ax2 =fig.add_axes([0.2,0.5,0.2 ,0.2])
ax2.plot(x,y ,color ='red')
```

```python
fig = plt .figure()
ax1 =fig.add_axes([0,0,1,1])
ax1.plot(x,z)
#ax1.xlimits(1)
#ax1.ylimits(1)
ax2 =fig.add_axes([0.2,0.5,.4,.4])
ax2.plot(x,y)
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

```python
fig,ax= plt.subplots(nrows=1 ,ncols=2 ,figsize=(8,8))

ax[0].plot(x,y ,lw =4 ,ls ='--' ,color ='pink')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('frist plots')
ax[1].plot(x,z,lw = 5 ,ls = '--', color= 'green')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_title('second plots')
plt.tight_layout()
```
