```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x= np.linspace(0 ,10 ,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x_Axis')
plt.ylabel('y_Axis')
plt.title('first fig')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
fig =plt.figure()
axes =fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x_Axis')
axes.set_ylabel('y_Axis')
axes.set_title('first fig')
```

```python
fig =plt.figure()
axes1 =fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2 ,0.4,0.4,0.3])
axes1.plot(x,y)
axes1.set_xlabel('x_Axis')
axes1.set_ylabel('y_Axis')
axes1.set_title('big fig')
axes2.plot(y,x)
axes2.set_xlabel('x_Axis')
axes2.set_ylabel('y_Axis')
axes2.set_title('small fig')
```

```python
fig,axes3= plt.subplots(nrows=1 ,ncols=2)

axes3[0].plot(x,y)
axes3[0].set_xlabel('x_Axis')
axes3[0].set_ylabel('y_Axis')
axes3[0].set_title('1 fig')
axes3[1].plot(y,x ,'g')
axes3[1].set_xlabel('x_Axis')
axes3[1].set_ylabel('y_Axis')
axes3[1].set_title('2 fig')
plt.tight_layout()
```

```python
fig,axes4= plt.subplots(nrows=2,ncols=1 ,figsize=(8,6))
axes4[0].plot(x,y)
axes4[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig()
```

```python
fig = plt.figure()
a =fig.add_axes([1,1,1,1])
a.plot(x,y,label='xy')
a.plot(y,x, label='yx')
a.legend(loc=0)
```

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
a1.plot(x,y ,color='green' , lw= 3 ,ls='-' ,marker ='*' ,ms =7
        ) #rgb hex code when can use 
a1.set_xlim([0,1])
a1.set_ylim([0,2])
```
